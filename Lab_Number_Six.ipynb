```python
from PIL import Image, ImageFilter

photo = Image.open('gerald_lilia.jpg') # открываю фото
max_size = (800, 600) # задаю переменную с ее размерами
size_photo = photo.thumbnail(max_size) # уменьшаю размер фото при помощи thumbnail

zatenenie = ImageFilter.Kernel((3, 3), [-1, -1, -1,-1, 9, -1, -1, -1, -1], scale=9, offset=5) # Делает небольшое размытие с затенением, повышает резкость
kontyr = ImageFilter.Kernel((3, 3), [-1, 0, 0, 0, 1, 0, 0, 0, 0], scale=9, offset=2) # Выводит черный контур фотографии

photo_filtered1 = photo.filter(zatenenie) # рименяю фильтр затенения
photo_filtered2 = photo.filter(kontyr) # применяю фильтр контура

photo_filtered1.show() # показываю фото 1 с фильтром затенения
photo_filtered2.show() # показываю фото 2 с фильтром контура
```

3. Что такое фильтрация изображений? Приведите примеры практического применения фильтров
 * Фильтрация изображений - это процесс обработки изображения с использованием математических операций для достижения определенного эффекта. Фильтры могут использоваться для улучшения качества изображения, устранения шума, увеличения резкости или создания художественных эффектов.
 * Улучшение качества изображения: Фильтры могут использоваться для коррекции цветового баланса, контрастности и яркости изображения.
 * Устранение шума: Фильтры могут использоваться для устранения шума на изображении, вызванного низким освещением или высокой чувствительностью камеры.
 * Увеличение резкости: Фильтры могут использоваться для увеличения резкости изображения, чтобы сделать его более четким и детализированным.
 * Художественные эффекты: Фильтры могут использоваться для создания художественных эффектов, таких как размытие, сепия или черно-белый режим.

4. Поясните формат изображения RGB.
 * RGB - это цветовой режим, используемый для представления цветных изображений на экранах компьютеров и телевизоров. RGB означает красный (Red), зеленый (Green) и синий (Blue), поскольку эти три цвета используются для создания всех остальных цветов.

5. Поясните значение терминов: растр, формат изображения, палитра, глубина, разрешение изображения.
 * Растр - это тип изображения, состоящий из пикселей, расположенных в сетке. Каждый пиксель имеет определенный цвет и яркость. Растровые изображения обычно используются для фотографий и других изображений с большим количеством деталей и градиентов.

 * Формат изображения - это стандарт, определяющий, как данные об изображении хранятся и кодируются в файле. Существует множество различных форматов изображений, таких как JPEG, PNG и GIF. Каждый формат имеет свои особенности и может быть лучше подходить для определенных типов изображений.

 * Палитра - это набор цветов, используемых в изображении. В некоторых форматах изображений, таких как GIF и PNG-8, палитра ограничена небольшим числом цветов (обычно 256 или меньше), и каждый пиксель изображения ссылается на один из цветов палитры.

 * Глубина - это количество битов данных, используемых для представления каждого пикселя изображения. Глубина определяет количество возможных цветов и оттенков серого в изображении. Например, 1-битное изображение может содержать только два цвета (обычно черный и белый), а 24-битное изображение может содержать миллионы различных цветов.

 * Разрешение изображения - это количество пикселей на единицу длины (обычно на дюйм) в изображении. Разрешение определяет четкость и детализацию изображения. Чем выше разрешение, тем больше деталей может быть отображено на изображении.
