# Выполнение задания

Загружаем свой csv файлик по варианту

```python
!wget https://raw.githubusercontent.com/nae-stormrage/python_ncfu/master/scripts/data/p4_data_15.csv
```

```python
import matplotlib.pyplot as plt
import csv

x = [] # Создали пустой список для х
y = [] # Создали пустой список для у

with open('p4_data_15.csv', 'r') as csvfile: # Открыли файл p4_data_15.csv
    plots = csv.reader(csvfile, delimiter=',') # Сохранили в переменную plots все данные из файла p4_data_15.csv
    next(plots) # Пропустили первую строку с заголовоком
    for row in plots:
        x_value = row[4].replace('$', '') # Убрали знак доллара в 4 колонке
        y_value = row[5].replace('$', '') # Убрали знак доллара в 5 колонке
        x.append(float(x_value)) # Добавляем значения в список х
        y.append(float(y_value)) # Добавляем значения в список y

# Создание линейного графика
plt.figure()
plt.plot(x)
plt.xlabel('x')
plt.title('Линейный график')

# Создание столбчатой диаграммы
plt.figure()
plt.bar(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Столбчатая диаграмма')

plt.show()
```

2. Продемонстрируйте варианты создания массива numpy: из списка, как диапазон с заданным шагом, массив нуле, единиц.

```python
# Создание массива NumPy из списка Python
import numpy as np
my_list = [1, 2, 3]
my_array1 = np.array(my_list)

#Создание массива NumPy как диапазона с заданным шагом:
my_array2 = np.arange(start=0, stop=10, step=2)

#Создание массива NumPy из нулей:
my_array3 = np.zeros(shape=(3, 3))

#Создание массива NumPy из единиц:
my_array4 = np.ones(shape=(3, 3))

print(my_array1, '\n')
print(my_array2, '\n')
print(my_array3, '\n')
print(my_array4, '\n')
```

3. Продемонстрируйте выполнение векторных операций с массивами numpy: умножение вектора на число, сложение векторов.

```python
#Умножение вектора на число:
import numpy as np

my_array = np.array([1, 2, 3])
result1 = my_array * 2

#Сложение векторов:
array_1 = np.array([1, 2, 3])
array_2 = np.array([4, 5, 6])
result2 = array_1 + array_2

print(result1, '\n')
print(result2, '\n')
```
