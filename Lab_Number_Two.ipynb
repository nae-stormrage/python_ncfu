## Задание

Напишите скрипт поиска чисел, кратных 7, но не кратных 5, в списке, состоящем из элементов произвольного типа.

## Решение задачи


Для начала вы сгенерируем список из различных типов данных. Затем мы создадим второй список, но пока пустой. В него мы будем записывать числа, которые удовлетворяют условию задачи. Составим простой цикл, который будет проверять каждый элемент нашего сгенерированного списка и проверять на удовлетворённость нашим условиям.


```python
f = [1327,"labirint",14, 21, 1753, 9.47, 1233, 101.4, 102, 28, 84, 35, 881, "Money", 416, 19.04, 1678, 194, "Love"]
p = []

for i in range(0, len(f)):
    e = f[i]

    if type(e) == int:
        if e%7 == 0 and e%5 !=0:
            p.append(e)
        else:
            print("Число:", e,"- Не удовлетворяет условиям.")
    else:
        print("Вовсе не является числом, либо не является целым числом")

print(p)
```
